```python
import urllib.request
import os
```

```python
cat = open('/Users/championlin/Desktop/kitty_cat.txt').readlines()
```

```python
def download_img():
    i = 0
    for url in cat:
        i += 1
        open('/Users/championlin/Desktop/tensorflow_tut/transfor_learning_img/cat%s.png' %i, 'wb')
        file_name = '/Users/championlin/Desktop/tensorflow_tut/transfor_learning_img/cat%s.png' %i
        try:
            urllib.request.urlretrieve(url=url, filename=file_name)
            print('download: ' + url)
        except:
            os.remove(file_name)
            print('no image')
    print('finished')
```

```python
download_img()
```

```python
for file in os.listdir('/Users/championlin/Desktop/tensorflow_tut/transfor_learning_img/'):
    root = '/Users/championlin/Desktop/tensorflow_tut/transfor_learning_img/'
    if os.path.getsize(root + file) <= 3 * 1024:
        os.remove(root + file)
        print('delete file:' + file)
print('finished.')
```
