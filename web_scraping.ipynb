## 博客來瀏覽書所推薦其他書之爬蟲

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen 
import re
import random
```

```python
html = urlopen('http://www.books.com.tw/products/0010754327?loc=P_asb_001').read().decode(utf-8)
print(html)
```

```python
soup = BeautifulSoup(html, features = 'lxml')
href = soup.find('a',{'href':re.compile('/products/(.{10})\Sloc\SP')})
print(href)
```

```python
print(href.img['alt'],'  url =', href['href'])
```

```python
his = []
sub_url = soup.find_all('a',{'href':re.compile('/products/(.{10})\Sloc\SP_asv')})
# print(sub_url[0]['href'])
if len(sub_url) != 0:
    his.append(random.sample(sub_url, 1)[0]['href'])
else:
    his.pop()
print(his)
```

```python
his = ['http://www.books.com.tw/products/0010754327?loc=P_asb_001']
for i in range(20):
    url = his[-1]
    url = urlopen(url).read().decode('utf-8')
    soup = BeautifulSoup(url, features = 'lxml')
    href = soup.find('a',{'href':re.compile('/products/(.{10})\Sloc\SP_asv')})
    print(i, href.img['alt'],'  url : ',href['href'])
    
    sub_url = soup.find_all('a',{'href':re.compile('/products/(.{10})\Sloc\SP_asv')})
    if len(sub_url) != 0:
        his.append(random.sample(sub_url, 1)[0]['href'])
    else:
        his.pop()
```
