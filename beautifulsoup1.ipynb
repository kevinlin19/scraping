因為使用正規表示式有時候仍然會有含有超連結的內容，故我們試著使用BeautifulSoup來替代正規表示式

```python
#導入module
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
#html
html = urlopen("https://dalab.ie.nthu.edu.tw/news_content.php?id=56").read().decode('utf-8')
print(html)
```

```python
soup = BeautifulSoup(html, features='lxml')
print(soup.h1)
```

```python
print(soup.p)
```

```python
print(soup.a)
```

```python
all_herf = soup.find_all('a')
print(all_herf)

for l in all_herf:
    print(l['href'])
```
